# DATA WRANGLING
On assessing the 'gun data', we observe that there is a huge amount of missing values on some culumns as compared to the amount of rows we have. So, we would drop this columns: drop columns with over 60% missing data and adjust the totals.

Fot the US census data, we would drop the fact_note column.

### Cleaning the gun data set

```python
# in a dataset of 12485, I will remove all columns with missing data greater that 7500
# This is for us to have a minimum row of 5000 to work with

for i in df_gun.columns:
    if missing_data[i] > 7500:
        df_gun = df_gun.drop(columns = [i], axis=1)
df_gun.head(2)
```

```python
# checking again for missing data after removing redundant columns
missing_data = df_gun.isnull().sum()
missing_data
```

```python
# Drop rows with null values
df_gun = df_gun.dropna()

# checking again for missing data
missing_data = df_gun.isnull().sum()
missing_data
```

```python
#convert the data type for the month column to date time
df_gun['month'] = pd.to_datetime(df_gun['month'])
df_gun.info()
```

#### Adjusting the total values after cleaning the gun data set

```python
df_gun['totals'] = df_gun.iloc[:, 2:-1].sum(axis=1)
df_gun['totals']
```

```python
df_gun.head(1)
```

### Search and tidy up duplicates and non unique values

```python
df_gun.duplicated().sum()
```

## Cleaning the US Census data set

```python
df_census.drop(df_census.index[64:], inplace=True)
df_census.tail(30)
```

### Convert strings to float and rename columns for the us census data

```python
# replace spaces with underscores and lowercase labels for us census dataset
df_census.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
df_census.columns
```

```python
df_census = df_census.drop('fact_note', axis=1)
df_census
```

#### Remove and replace special characters $ % ,

```python
# removing symbols like $, D, , and % sign from the values of the data set
df_census = df_census.replace({'\$':'', ',':'', '%':'', 'D':' ' , 'Z':' ' , 'FN':' ' }, regex=True)

#replace empty space with NaN
df_census = df_census.replace(r'^\s*$', np.NaN, regex=True)
df_census.tail(30)
```

#### Convert objects to floats

```python
for i in df_census.columns:
    if i == 'fact':
        pass
    else:
        df_census[i] = df_census[i].astype(float)
df_census.info()
```

```python
df_census.isnull().sum()
```

#### Fill missing data with the mean of each row

```python
# Find the mean of each row and fill in the NaN values with the mean
mean_val = df_census.iloc[1:50].mean(axis=1) 
df_census.fillna(mean_val, inplace = True)
df_census.isnull().sum()
```

```python
# Find the duplicate rows
df_census.duplicated().sum()
```

```python
df_census.info()
```

# DRAW CONCLUSIONS

#### What census data is associated with high census per capital?

```python
"""
To get the census data associated with high census per capital, I will divide the combined dataset (df_gun_census)
into high, mid and low using the per_capital_income column. Then extract the high census per capital values
"""
df_gun_census['per_capital_income'].describe()
```

```python
# creating a categorical variable data of high, mid, low per_capital_income
bin_edges = [21057, 25443.75, 27669.50, 30977.00, 38803.00] #increasing bin monotolitically
bin_names = ['low', 'mid', 'mid_high', 'high']
df_gun_census['income_level'] = pd.cut(df_gun_census['per_capital_income'], bin_edges, labels=bin_names)

df_gun_census.head(3)
```

```python
# List of states, population, employment rate,gun purchase associated with high per capital income
df_gun_census.query('income_level == "high"')
```

```python
#Census data associated with high per capital income earning states
df_gun_census.query('income_level == "high"').describe()
```

```python
df_income_level = df_gun_census.groupby('income_level', as_index=False)['per_capital_income', 'total_public_gun'].sum()
df_income_level
```

```python
df_income_level.plot(x='income_level', kind='bar', figsize=(8,8), title= 'Total Public Gun Sales Per capital Income');
```

#### Which state have had the highest growth in gun registration?

```python
"""
Group by year and by state the total permit that was issued
"""
df_permit2010 = df_gun.query("month == '2010-01-01'").groupby('state', as_index=False)['permit'].sum()
df_permit2010.head()
```

```python
df_permit2017 = df_gun.query("month == '2017-09-01'").groupby('state', as_index=False)['permit'].sum()
df_permit2017.head()
```

```python
df_permit_2010_2017 = df_permit2010.merge(df_permit2017, left_on = 'state', right_on = 'state', how= 'inner')
df_permit_2010_2017.head()
```

```python
# subtract permit_x from permit_y to get the increase in gun registration
df_permit_2010_2017['permit_increase'] = df_permit_2010_2017['permit_y'] - df_permit_2010_2017['permit_x']
df_permit_2010_2017.head()
```

```python
# state with highest registration using bar chart
df_permit_2010_2017.plot(x='state', y='permit_increase', kind='barh', figsize=(12,12), title= 'Permit increase Per State');
```
