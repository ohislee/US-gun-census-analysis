# DATA WRANGLING
On assessing the `gun_data.xlsx` and the `us_census_data.csv`, we observed that there is a huge amount of missing values on some culumns for the gun data as compared to the amount of rows we have. So, we would drop these columns. I would be using a for loop to drop columns with over `7500` missing values. 

Fot the US census data, we would drop the fact_note column and write ups at the base of the dataset would be removed.

## Cleaning the gun data set

```python
# in a dataset of 12485, I will remove all columns with missing data greater that 7500
# This is for us to have a minimum row of 5000 to work with

for i in df_gun.columns:
    if missing_data[i] > 7500:
        df_gun = df_gun.drop(columns = [i], axis=1)
df_gun.head(2)
```

```python
# checking again for missing data after removing redundant columns
missing_data = df_gun.isnull().sum()
missing_data
```

```python
# Drop rows with null values
df_gun = df_gun.dropna()

# checking again for missing data
missing_data = df_gun.isnull().sum()
missing_data
```

```python
# convert the data type for the month column to date time
df_gun['month'] = pd.to_datetime(df_gun['month'])
df_gun.info()
```

## Cleaning the US Census data set

### Convert strings to float and rename columns for the us census data

```python
# Lets define a function for renaminng columns
def rename_column(cols):
    
    # replace spaces with underscores and lowercase labels for us census dataset
    cols.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
    
    return cols.columns # return new names of columns
```

```python
# replace spaces with underscores and lowercase labels for us census dataset
rename_column(df_census)
```

```python
# There are a lot of missing values in the fact_note column, lets drop it
df_census = df_census.drop('fact_note', axis=1)
df_census
```

#### Remove and replace special characters $ % ,

```python
# removing symbols like $, D, , and % sign from the values of the data set
df_census = df_census.replace({'\$':'', ',':'', '%':'', 'D':' ' , 'Z':' ' , 'FN':' ' }, regex=True)

#replace empty space with NaN
df_census = df_census.replace(r'^\s*$', np.NaN, regex=True)
df_census.tail(30)
```

#### Convert objects to floats for census data

```python
# Lets convert all column data type to float except for the fact column on the us census data
for i in df_census.columns:
    if i == 'fact':
        pass
    else:
        df_census[i] = df_census[i].astype(float)
df_census.info()
```

```python
# checking for missing values on the us census data
df_census.isnull().sum()
```

```python
# Find the mean of each row and fill in the NaN values with the mean
mean_val = df_census.iloc[1:50].mean(axis=1) 
df_census.fillna(mean_val, inplace = True)
df_census.isnull().sum()
```

### Lets create a function to visualize relationship between variables using seaborn joint plot

```python
def joint_plot(df, xVar, yVar):
    """
    This is a function to define the relationship between 
    two variables using seaborn jointplot
    """
    g = sns.jointplot(x = xVar, y = yVar, data = df, kind = 'reg', line_kws = {'color':'maroon'});
    
    # tidying up labels for x and y axis
    xVar = xVar.replace('_',' ').title()
    yVar = yVar.replace('_',' ').title()
    
    # add main title to the plot
    plt.suptitle(f"Relationship between {yVar} and {xVar}".title(), weight = 'bold', y=1.04);
    
    # set axis labels
    g.set_axis_labels(xVar, yVar)
    
    # set the fig size
    g.fig.set_size_inches(14,6);
```

#### Which state have had the highest growth in gun registration?
To answer this question, we have to query the permit column to get the total permit issued for each state on the first month of our df_gun dataset and also the total permit issued on the last month. The values gotten for each state on the first month would be substracted from that of the last month. The maximum value gotten would indicate the state that has gotten the highest increase in gun registration.
> The state with the highest growth in gun registration is **Kentucky** as shown by the graph below.
>- The horizontal axis represent states while the y-axis represent permit increase.
>- States like Illinoise and Utah have their number of gun registration drop.

```python
# Lets query the data to get all the gun registration done on the first month and group by states
df_permit2010 = df_gun.query("month == '2010-01-01'").groupby('state', as_index=False)['permit'].sum()
df_permit2010.head()
```

```python
# Lets query the data to get all the gun registration done on the last month and group by states
df_permit2017 = df_gun.query("month == '2017-09-01'").groupby('state', as_index=False)['permit'].sum()
df_permit2017.head()
```

```python
# Next, we merge the first month and the last month data set
df_permit_2010_2017 = df_permit2010.merge(df_permit2017, left_on = 'state', right_on = 'state', how= 'inner')
df_permit_2010_2017.head()
```

```python
# subtract permit_x from permit_y to get the increase in gun registration
df_permit_2010_2017['permit_increase'] = df_permit_2010_2017['permit_y'] - df_permit_2010_2017['permit_x']
df_permit_2010_2017.head()
```

```python
# The state with the highest increase in gun registration would be gotten by finding max of permit_increase
df_permit_2010_2017[df_permit_2010_2017.permit_increase == df_permit_2010_2017.permit_increase.max()]
```

```python
# We can also view the highest increase in gun registration using a bar chart
df_permit_2010_2017.plot(x='state', 
                         y='permit_increase', 
                         kind='bar', figsize=(14, 6), 
                         title= 'Permit increase Per State')

plt.ylabel('Increase In Permit')
plt.xlabel('States');
```
